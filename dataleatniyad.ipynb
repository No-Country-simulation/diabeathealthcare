```python
import csv
import random
import pandas as pd
import numpy as np

# Funciones auxiliares para generar datos aleatorios de niños y adolescentes
def generar_nombre(genero):
    nombres_hombres = ["Alberto", "Bernardo", "Carlos", "Daniel", "Eduardo", "Fernando", "Gustavo", "Hector", "Ignacio", "Javier",  "Luis", "Manuel", "Nicolas", "Oscar", "Pablo", "Quintin", "Ricardo", "Sergio", "Tomas", "Ulises", "Victor", "Walter",  "Andres", "Braulio", "Cristian", "David", "Esteban", "Fabian", "Gabriel", "Hugo", "Ivan", "Jose", "Leonardo", "Marcos",  "Pedro",  "Raul", "Samuel", "Valentin"]
    nombres_mujeres = ["Adriana",  "Alejandra", "Alicia", "Alma",  "Ana", "Andrea", "Angela",  "Beatriz", "Camila", "Carla", "Carmen", "Carolina", "Clara", "Claudia", "Daniela", "Diana", "Elena", "Elisa", "Emilia", "Esperanza",  "Fabiola", "Gabriela", "Gloria", "Helena", "Ingrid", "Irene", "Isabel", "Laura", "Lucia", "Luisa", "Maria", "Marta", "Natalia", "Noelia", "Olga", "Patricia", "Raquel", "Rosa"]
    return random.choice(nombres_hombres if genero == "Hombre" else nombres_mujeres)

def generar_apellido():
    apellidos = ["Garcia", "Martinez", "Lopez", "Hernandez", "Gonzalez", "Perez", "Rodriguez", "Sanchez", "Ramirez", "Cruz", "Flores", "Gomez", "Morales", "Diaz", "Vasquez", "Torres", "Vargas", "Jimenez", "Reyes", "Ortiz", "Ruiz", "Castillo", "Chavez", "Ramos", "Mendoza", "Fernandez", "Molina", "Silva", "Rojas", "Guerrero", "Castro", "Romero", "Soto", "Dominguez", "Ortega", "Delgado", "Pena", "Campos", "Medina", "Nunez"]
    return random.choice(apellidos)

def generar_genero():
    return random.choice(["Hombre", "Mujer"])

def generar_peso(genero, edad):
    if edad <= 2:
        return random.uniform(3, 15)  # Peso en kg para bebés y niños pequeños
    elif edad <= 12:
        return random.uniform(15, 50)  # Peso en kg para niños
    else:
        return random.uniform(40, 80)  # Peso en kg para adolescentes

def generar_altura(genero, edad):
    if edad <= 2:
        return random.uniform(50, 100)  # Altura en cm para bebés y niños pequeños
    elif edad <= 12:
        return random.uniform(100, 150)  # Altura en cm para niños
    else:
        return random.uniform(140, 190)  # Altura en cm para adolescentes

def generar_edad():
    return random.randint(0, 18)

# Generar datos para los usuarios niños y adolescentes
usuarios = [["Nombre", "Genero", "Peso", "Altura", "Edad"]]

for i in range(100):
    genero = generar_genero()
    edad = generar_edad()
    nombre = generar_nombre(genero) + " " + generar_apellido()
    peso = generar_peso(genero, edad)
    altura = generar_altura(genero, edad)
    usuarios.append([nombre, genero, peso, altura, edad])

# Crear y escribir en el archivo CSV
with open('Usuarios_ninos.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(usuarios)

print("Archivo CSV con datos de niños y adolescentes creado con éxito.")

# Ahora vamos a agregar las columnas adicionales al archivo CSV generado
# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv('Usuarios_ninos.csv')

# Generar datos adicionales
num_filas = len(df)

data_adicional = {
    'Actividad Fisica': np.random.randint(1, 4, num_filas),  # Nivel de actividad fisica entre 1 y 3
    'Ecuacion': np.random.choice(['Mifflin', 'Harris'], num_filas),
    'Objetivos': np.random.choice(['Crecimiento saludable', 'Mantenimiento del peso', 'Aumento de masa muscular', 'Mejorar la actividad física'], num_filas),
    'Recomendacion': np.random.choice(['Comer más frutas y verduras', 'Aumentar proteínas', 'Reducir azúcares', 'Hacer ejercicio regularmente'], num_filas),
    'Calorias consumidas por dia': np.random.uniform(1000, 2500, num_filas).round(1),  # Calorias consumidas entre 1000 y 2500
    'Carbohidratos': np.random.uniform(100, 300, num_filas).round(1),  # Carbohidratos entre 100 y 300 g
    'Proteinas': np.random.uniform(30, 150, num_filas).round(1),  # Proteinas entre 30 y 150 g
    'Lipidos': np.random.uniform(20, 80, num_filas).round(1)  # Lipidos entre 20 y 80 g
}

# Calcular el requerimiento calorico diario
def calcular_requerimiento(row):
    if row['Ecuacion'] == 'Mifflin':
        if row['Genero'] == 'Hombre':
            return 10 * row['Peso'] + 6.25 * row['Altura'] - 5 * row['Edad'] + 5
        else:
            return 10 * row['Peso'] + 6.25 * row['Altura'] - 5 * row['Edad'] - 161
    else:  # Harris
        if row['Genero'] == 'Hombre':
            return 88.362 + 13.397 * row['Peso'] + 4.799 * row['Altura'] - 5.677 * row['Edad']
        else:
            return 447.593 + 9.247 * row['Peso'] + 3.098 * row['Altura'] - 4.330 * row['Edad']

# Agregar los datos adicionales al DataFrame
for col, values in data_adicional.items():
    df[col] = values

# Calcular la columna del requerimiento calorico diario
df['Requerimiento (Kcal diarias)'] = df.apply(calcular_requerimiento, axis=1).round(1)

# Guardar el DataFrame en un archivo CSV actualizado
df.to_csv('Usuarios_ninos_actualizado.csv', index=False)

print("Archivo CSV actualizado con datos adicionales de niños y adolescentes creado con éxito.")

# Cargar recetas desde el archivo CSV
recipes_df = pd.read_csv('recipes.csv')

# Filtrar recetas adecuadas para niños y adolescentes (asumimos que las recetas con menos de 500 KCal son adecuadas)
recipes_adecuadas_df = recipes_df[recipes_df['KCal'] < 500]

# Crear una lista de días de la semana
dias_semana = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Crear un nuevo DataFrame para almacenar los datos extendidos
extended_data = []

# Asignar aleatoriamente recetas adecuadas a cada usuario niño o adolescente para cada día de la semana
for index, user in df.iterrows():
    for dia in dias_semana:
        desayuno = recipes_adecuadas_df.sample().iloc[0]
        almuerzo = recipes_adecuadas_df.sample().iloc[0]
        cena = recipes_adecuadas_df.sample().iloc[0]

        extended_data.append({
            'Nombre': user['Nombre'],
            'Genero': user['Genero'],
            'Peso': user['Peso'],
            'Altura': user['Altura'],
            'Edad': user['Edad'],
            'Actividad Fisica': user['Actividad Fisica'],
            'Ecuacion': user['Ecuacion'],
            'Objetivos': user['Objetivos'],
            'Recomendacion': user['Recomendacion'],
            'Calorias consumidas por dia': user['Calorias consumidas por dia'],
            'Carbohidratos': user['Carbohidratos'],
            'Proteinas': user['Proteinas'],
            'Lipidos': user['Lipidos'],
            'Desayuno': desayuno['food'],
            'Almuerzo': almuerzo['food'],
            'Cena': cena['food'],
            'Desayuno KCal': desayuno['KCal'],
            'Almuerzo KCal': almuerzo['KCal'],
            'Cena KCal': cena['KCal'],
            'Dia': dia,
            'Requerimiento (Kcal diarias)': user['Requerimiento (Kcal diarias)']
        })

# Crear un DataFrame con los datos extendidos
extended_df = pd.DataFrame(extended_data)

# Guardar el DataFrame extendido en un nuevo archivo CSV
extended_df.to_csv('Usuarios_ninos_comidas_semana.csv', index=False)

print("Archivo CSV con datos extendidos de comidas para niños y adolescentes creado con éxito.")
```
